```python
import nltk
import re
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def clean_text(text):
    # Minuscolo
    text = text.lower()
    # Rimuovi numeri e punteggiatura
    text = re.sub(r'[^a-z\s]', '', text)
    # Rimuovi stopwords
    words = text.split()
    words = [w for w in words if w not in stop_words and len(w) > 2]
    return ' '.join(words)

texts_clean = {}
for company, text in texts.items():
    texts_clean[company] = clean_text(text)
    print(f'✓ {company} — {len(texts_clean[company].split())} parole')

print(f'\nTotale aziende pulite: {len(texts_clean)}')
```

```python
import numpy as np

# Keyword ESG rilevanti
ESG_KEYWORDS = [
    'emission', 'carbon', 'climate', 'sustainab', 'governance', 
    'diversity', 'target', 'commitment', 'environment', 'social',
    'renewable', 'waste', 'water', 'energy', 'safety', 'human rights',
    'net zero', 'scope', 'transparency', 'accountability'
]

def has_esg_keyword(chunk):
    chunk_lower = chunk.lower()
    return any(kw in chunk_lower for kw in ESG_KEYWORDS)

def analyze_sentiment_smart(text, max_chunks=150):
    words = text.split()
    all_chunks = [' '.join(words[i:i+400]) for i in range(0, len(words), 400)]
    
    # Filtra chunk con keyword ESG
    esg_chunks = [c for c in all_chunks if has_esg_keyword(c)]
    
    # Se troppi, campiona in modo stratificato lungo il documento
    if len(esg_chunks) > max_chunks:
        indices = np.linspace(0, len(esg_chunks)-1, max_chunks, dtype=int)
        esg_chunks = [esg_chunks[i] for i in indices]
    
    positive_scores = []
    negative_scores = []
    neutral_scores = []
    
    for chunk in esg_chunks:
        if len(chunk.strip()) == 0:
            continue
        result = finbert(chunk[:512])[0]
        for item in result:
            if item['label'] == 'positive':
                positive_scores.append(item['score'])
            elif item['label'] == 'negative':
                negative_scores.append(item['score'])
            elif item['label'] == 'neutral':
                neutral_scores.append(item['score'])
    
    return {
        'positive': sum(positive_scores) / len(positive_scores) if positive_scores else 0,
        'negative': sum(negative_scores) / len(negative_scores) if negative_scores else 0,
        'neutral': sum(neutral_scores) / len(neutral_scores) if neutral_scores else 0,
        'chunks_analyzed': len(esg_chunks)
    }

sentiment_results = {}
for company, text in texts_clean.items():
    print(f'Analisi: {company}...')
    sentiment_results[company] = analyze_sentiment_smart(text)
    r = sentiment_results[company]
    print(f'  → positive: {r["positive"]:.3f} | negative: {r["negative"]:.3f} | neutral: {r["neutral"]:.3f} | chunks: {r["chunks_analyzed"]}')

print('\nAnalisi completata!')
```
